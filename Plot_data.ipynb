```python
# importing resources
import pandas as pd
```

```python
import seaborn as sns
```

```python
# load seaborn dataset
tips = sns.load_dataset('tips')
```

```python
import matplotlib.pyplot as plt
```

```python
# plotting histogram for the data in 'tip' column of the tips df
tips.tip.plot(kind='hist')
```

```python
plt.show() # no need if working in Jupyter Notebook
```

```python

tips.tip.plot(kind='hist')
plt.show()
```

```python
# value_counts calculates number of each reponse under a column
# e.g. here in this datset, 15 people dont smoke and 93 of them smoke
# assigning variable name to the output
cts = tips.smoker.value_counts()
```

```python
# looking at the value_counts() variable
cts
```

```python
# plotting the value_counts() variable
cts.plot(kind='bar')
```

```python
# we can also use countplot to do the same thing
sns.countplot(x='smoker', data=tips)
```

```python
# distribution plot for total_bill data in tips df
sns.distplot(tips.total_bill)
```

```python
# linear regression plot for relationship between tip and total_bill
sns.lmplot(x='total_bill', y = 'tip', data=tips)
```

```python
# linear regression plot for relationship between tip and total_bill
# colored according to sex using 'hue' option
sns.lmplot(x='total_bill', y = 'tip', data=tips, hue = 'sex')
```

```python
# regression plot with no regression line
sns.lmplot(x='total_bill', y = 'tip', data=tips, hue = 'sex', fit_reg=False)
```

```python
# fitting regression in plots seperated into columns based on 'smoker' column (yes or no)
sns.lmplot(x='total_bill', y = 'tip', data=tips, hue = 'sex', fit_reg=False, col = 'smoker')
```

```python
# fitting regression in plots seperated into columns based on smoker
# and rows based on days of the week
sns.lmplot(x='total_bill', y = 'tip', data=tips, hue = 'sex', 
           fit_reg=False, col = 'smoker', row='day')
```

```python
# Can use facet grid to make grids

# the following line builds the matrix of empty plots
facet = sns.FacetGrid(tips, col = 'time', row = 'smoker', hue = 'sex')

# the following line fills the plots with the kind of user-defined plot and the data to be used
facet.map(plt.scatter, 'total_bill', 'tip')
```

```python
# using subplots with figure and axes defined
# defined axes can then be used to fill the plot
fig, (ax1) = plt.subplots(1,1)
ax1.scatter(tips.tip, tips.total_bill)
```

```python
# subplots with 1 row and 2 columns
# 2 axes for the two plots - we will use matplotlib pots here
# first axes is used to fill scatter plot in first figure
# 2nd axes is used to fill histogram plot in 2nd figure
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.scatter(tips.tip, tips.total_bill)
ax2.hist(tips.total_bill)
```

```python
# distribution plot
dis = sns.distplot(tips.total_bill)
```

```python
# now we will fill the subplots with seaborn plots
fig, (ax1, ax2) = plt.subplots(1, 2)
sns.distplot(tips.tip, ax=ax1)
sns.regplot(x='total_bill', y='tip', data = tips, ax=ax2)
```
